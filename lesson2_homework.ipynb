```python
# 生成站点位置和站点编码与名称对应关系以及其连接关系
from collections import defaultdict
station_connect = defaultdict(list)
station_points = {}
station_infs = {}
for i in range(len(json_geod['l'])):
    lines_name = json_geod['l'][i]['ln']
    lines_code = json_geod['l'][i]['ls']
    for j in range(len(json_geod['l'][i]['st'])-1):
        station_name = json_geod['l'][i]['st'][j]['n']
        next_station_name = json_geod['l'][i]['st'][j+1]['n']
        station_sign = json_geod['l'][i]['st'][j]['si']
        next_station_sign = json_geod['l'][i]['st'][j+1]['si']
        station_lng = json_geod['l'][i]['st'][j]['sl'].split(',')[0]
        next_station_lng = json_geod['l'][i]['st'][j+1]['sl'].split(',')[0]
        station_lat = json_geod['l'][i]['st'][j]['sl'].split(',')[1]
        next_station_lat = json_geod['l'][i]['st'][j+1]['sl'].split(',')[1]

        if station_name not in station_points:
            station_points[station_name] = (float(station_lng), float(station_lat))
        if station_sign not in station_infs:
            station_infs[station_sign] = station_name
        if next_station_name not in station_points:
            station_points[next_station_name] = (float(next_station_lng), float(next_station_lat))
        if next_station_sign not in station_infs:
            station_infs[next_station_sign] = next_station_name
        
        if station_name != next_station_name:
            if next_station_name not in station_connect[station_name]:
                station_connect[station_name].append(next_station_name)
            if station_name not in station_connect[next_station_name]:
                station_connect[next_station_name].append(station_name)
```

```python
# 计算两个地铁站之间的距离
from math import radians,sin,cos,asin,sqrt
def geodistance(station1, station2):
    
    lng1 ,lat1 = station_points[station1]
    lng2 ,lat2 = station_points[station2]
    
    lng1 = float(lng1)
    lat1 = float(lat1)
    lng2 = float(lng2)
    lat2 = float(lat2)
    
    lng1, lat1, lng2, lat2 = map(radians, [lng1, lat1, lng2, lat2])
    dlon=lng2-lng1
    dlat=lat2-lat1
    a=sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2 
    dis=2*asin(sqrt(a))*6371*1000
    return dis
```

```python
# 添加排序
# 最少站
def stations_less(paths):
    return sorted(paths, key=len)

# 最多站
def stations_as_much_possible(paths):
    return sorted(paths, key=len, reverse=True)

# 距离最小
def distance_less(paths):
    if len(paths) <= 1 : return paths
    
    
    def geo_distance(sub_path):
        distance = 0
        for i in range(1,len(sub_path)):
            distance += geodistance(sub_path[i], sub_path[i-1])
            
        return distance
    
    return sorted(paths, key=geo_distance)

# 距离最长
def distance_as_much_as_possible(paths):
    if len(paths) <= 1 : return paths
    
    
    def geo_distance(sub_path):
        distance = 0
        for i in range(1,len(sub_path)):
            distance += geodistance(sub_path[i], sub_path[i-1])
            
        return distance
    
    return sorted(paths, key=geo_distance, reverse=True)
```

```python
def search(start, destination, station_connect,sort_candidate):
    
    if start not in station_connect : return None
    if destination not in station_connect : return None
    
    visited = set()
    
    paths = [[start]]
    
    while paths:
        path = paths.pop(0)
        tmp_path = path[-1]
        
        if tmp_path in visited : continue
        
        for station in station_connect[tmp_path]:
            
            if station in path : continue
                
            new_path = path + [station]
            
            paths.append(new_path)
            
            if station == destination: return new_path
            
        visited.add(tmp_path)
        
        paths = sort_candidate(paths)
```

```python
# 深度优先
def dfs(start, destination, station_connect):
    
    if start not in station_connect : return None
    if destination not in station_connect : return None
    
    visited = set()
    
    paths = [[start]]
    
    while paths:
        path = paths.pop(0)
        tmp_path = path[-1]
        
        if tmp_path in visited : continue
        
        for station in station_connect[tmp_path]:
            
            if station in path : continue
                
            new_path = path + [station]
            
            paths.append(new_path)
            
            if station == destination: return new_path
            
        visited.add(tmp_path)
```

```python
# 广度优先
def bfs(start, destination, station_connect):
    
    if start not in station_connect : return None
    if destination not in station_connect : return None
    
    visited = set()
    
    paths = [[start]]
    
    while paths:
        path = paths.pop(0)
        tmp_path = path[-1]
        
        if tmp_path in visited : continue
        
        for station in station_connect[tmp_path]:
            
            if station in path : continue
                
            new_path = path + [station]
            
            paths = [new_path] + paths
            
            if station == destination: return new_path
            
        visited.add(tmp_path)
```

```python
# loss
def loss(y , y_predict):
    return sum((y_hat - y_predict_hat)**2 for y_hat, y_predict_hat in zip(list(y), list(y_predict))) / len(list(y))
```

```python
try_times = 5000

min_loss = float('inf')
best_k = random.random() * 200 - 100
best_b = random.random() * 200 - 100

direction = [
    (+1,+1),
    (+1,-1),
    (-1,+1),
    (-1,-1)
]

scalar = 0.1

new_direction = random.choice(direction)

for i in range(try_times):
    
    k_direction, b_direction = new_direction
    
    current_k = best_k + k_direction * scalar
    current_b = best_b + b_direction * scalar
    
    price_for_rand_k_and_b = [price(rm , current_k, current_b) for rm in X]
    
    predict_loss = loss(y, price_for_rand_k_and_b)
    
    if min_loss > predict_loss:
        min_loss = predict_loss
        best_k, best_b = current_k, current_b
        new_direction = new_direction
        print("current_time : {},current min_loss: {}, b_k : {}, b_b : {}".format(i, min_loss, best_k, best_b))
    else:
        new_direction = random.choice(direction)
```

# 梯度下降

```python
# K和B的导数
def partial_k(x, y , y_hat):
    n = len(y)
    gradient = 0
    gradient = sum((y_i - y_hat_i) * x_i for y_hat_i, y_i, x_i in zip(y_hat, y, x))
    return - 2 / n * gradient

def partial_b(y, y_hat):
    n = len(y)
    gradient = 0
    gradient = sum((y_i - y_hat_i) for y_hat_i, y_i in zip(y_hat, y))
    return - 2 / n * gradient
```

```python
try_times = 2000

min_loss = float('inf')
best_k = random.random() * 200 - 100
best_b = random.random() * 200 - 100

learning_rate = 10e-4


for i in range(try_times):
    
    price_for_rand_k_and_b = [price(rm , best_k, best_b) for rm in X]
    
    predict_loss = loss(y, price_for_rand_k_and_b)
    
    if min_loss > predict_loss:
        min_loss = predict_loss
        
        print("current_time : {},current min_loss: {}, b_k : {}, b_b : {}".format(i, min_loss, best_k, best_b))
    
    k_gradient = partial_k(X, y, price_for_rand_k_and_b)
    b_gradient = partial_b(y, price_for_rand_k_and_b)
    
    best_k = best_k + ( -1 * k_gradient) * learning_rate
    best_b = best_b + ( -1 * b_gradient) * learning_rate
```

# Answer following questions

1. Why do we need machine learning methods instead of creating a complicated formula?

Ans:复杂的公式通常情况不能得到，并且很耗费资源和时间，就算得出复杂公式之后泛化能力不一定很好。而利用计算机的强大计算功能通过机器学习方法，让机器通过规定的方式进行参数推算，得出较接近数据的拟合算式。

2. Wha't's the disadvantages of the 1st Random Choosen methods in our course?

Ans：通过随机的方式进行算式参数的拟合和计算，有点赌博成份，结果不可控，运算次数也不可控。并且随机参数的范围也不可控，需要猜测

3. Is the 2nd method supervised direction better than 1st one? What's the disadvantages of the 2nd supversied directin method?

Ans：2nd-method是在1st-method的基础上，根据loss函数的变化方法，指定参数的变化方向，使得loss向着更优化（更小）的方向继续变化。
但由于2nd-method在方向上也存在随机性，可能会出现loss在变化的方向上大小大小这样不断地循环更达不到最好的效果。

4. Why do we use Derivative / Gredient to fit a target function?

Ans ： 函数的导数反映的是函数变化的方向，使用导数方法去拟合目标函数可更准备和快速的确认参数调整的变化。

5. In the words 'Gredient Descent', what's the Gredient and what's the Descent?

Ans ： 梯度是指在空间上，函数因某参数或变量的变化更随之发生的变化（函数值最终结果的差如阶梯式的变化）
下降是指变化量的减少，就是函数导数中值趋向0。

6. What's the advantages of the 3rd gradient descent method compared to the previous methods?

Ans ： 较于之前的1st-method和2nd-method，3rd-method的处在与通过导数的方式，确认下一次参数的变化方向，可确认每一次的变化都是向着loss更小的方向进行的


7. Using the simple words to describe: What's the machine leanring.

Ans ： 机器学习就是利于计算机的强大计算功能，计算机可自行根据给定的变化拟合规则，通过不断的迭代计算，得出与目标函数相接近的函数参数。
